#### Функция:
$$\frac { x } {1+ x^4} $$
Отрезок:[0, 1]


```python
import pandas as pd
import numpy as np
from numpy import linalg
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(-50, 50, 50)                   
y = x /(1+x**4 )

plt.style.use('classic')                        
fig, ax = plt.subplots()   
ax.set_title('График', fontsize = 15)
ax.plot(x, y, color="green", label="функция - y(x)")     
ax.set_xlabel("x", fontsize = 12)                              
ax.set_ylabel("Значения функции", fontsize = 10)                             
ax.legend()                                    

plt.show()
```

#### Минимальное, максимальное и среднее значения¶

```python
a=min(y)
b=max(y)
print('min_value=', a)
print('max_value=', b)
c=(a+b)/2
print('average_value=',c)
```

### Производная функции

```python
def diff(x, y):
    len_arr = len(y)
    if len(x) != len_arr:
        return None
    array_diff = np.zeros(len_arr)
    for i in range(len_arr-1):
        array_diff[i] = (y[i+1] - y[i])/(x[i+1]-x[i])
    array_diff[len_arr-1] = array_diff[len_arr - 2]
    return array_diff
        
Dif = diff(y,x)

plt.style.use('classic')    
fig, ax = plt.subplots()
ax.set_title('Производная функции', fontsize = 15)
ax.set_xlabel("x", fontsize = 12)
ax.set_ylabel("Производная", fontsize = 10)
ax.plot(x,Dif, color="green")

plt.show()
```
